## Energia e potÃªncia de um sinal <a class="anchor" id="section_2_4"></a>

### Tamanho de um Sinal <a class="anchor" id="section_2_4_1"></a>

O tamanho de qualquer entidade Ã© um nÃºmero que indica a largura ou o comprimento da entidade. Genericamente falando, a amplitude do sinal varia com o tempo. **Como um sinal que existe em um certo intervalo de tempo com amplitude variante pode ser medido por um nÃºmero que irÃ¡ indicar o tamanho ou a forÃ§a do sinal?** Tal
medida deve considerar nÃ£o apenas a amplitude do sinal, mas tambÃ©m sua duraÃ§Ã£o. 

Por exemplo, se quisermos utilizar um Ãºnico nÃºmero V como medida do tamanho de um ser humano, devemos considerar nÃ£o somente seu peso, mas tambÃ©m sua altura. Se fizermos uma consideraÃ§Ã£o que a forma da pessoa Ã© um cilindro cuja variÃ¡vel Ã© o raio r (o qual varia com a altura h), entÃ£o uma possÃ­vel medida do tamanho de uma pessoa de altura H Ã© o volume V da pessoa, dado por $$V = \int_o^H r^2(h)dh$$

### Energia de um Sinal <a class="anchor" id="section_2_4_2"></a>

Argumentando desta forma, podemos considerar a Ã¡rea abaixo do sinal $x(t)$ como uma possÃ­vel medida de seu tamanho, pois a Ã¡rea irÃ¡ considerar nÃ£o somente a amplitude, mas tambÃ©m sua duraÃ§Ã£o.$$Tamanho = \int_{-\infty}^{\infty} x(t)dt$$ 


```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

tempo = np.linspace(-10, 10, 1000)
amplitude = np.sin(x)

plt.fill_between(tempo, amplitude, step="pre", alpha=0.4)
plt.plot(tempo, amplitude, drawstyle="steps")
plt.xlabel('tempo (t)')
plt.ylabel('Amplitude  x(t)')

plt.show()
```

Entretanto, esta medida ainda Ã© defeituosa, pois mesmo para um sinal grande $x(t)$, suas Ã¡reas positivas e negativas podem se cancelar, indicando um sinal de tamanho pequeno:

Esta dificuldade pode ser corrigida pela definiÃ§Ã£o do tamanho do sinal como a Ã¡rea debaixo de $x^2(t)$ a qual Ã© sempre positiva. Podemos chamar essa medida de **energia do sinal**: $$E = \int_{-\infty}^{\infty} x^2(t)dt\$$

```python
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

x = np.linspace(-10, 10, 1000)
y1 = np.power(np.sin(x),2)

plt.fill_between(x, y1, step="pre", alpha=0.4)
plt.plot(x, y1, drawstyle="steps")
plt.xlabel('tempo (t)')
plt.ylabel('Amplitude  x(t)')


plt.show()
```

Essa definiÃ§Ã£o pode ser generalizada para um sinal complexo $x(t)$, sendo dada por $$E = \int_{-\infty}^{\infty} |x(t)|^2 dt$$

**A energia do sinal deve ser finita** para que seja uma medida significativa do tamanho do sinal. 
Uma condiÃ§Ã£o necessÃ¡ria para que a energia seja finita Ã© que a amplitude do sinal â 0 quando |t|â â. Caso contrÃ¡rio a integral da equaÃ§Ã£o acima nÃ£o irÃ¡ convergir.



### Potencia de um sinal  <a class="anchor" id="section_2_4_3"></a>

Quando a amplitude do sinal x(t) nÃ£o â 0 quando |t|ââ, a energia do sinal Ã© infinita. 



Uma medida mais significativa do tamanho do sinal neste caso Ã© a energia mÃ©dia, se ela existir. Esta medida Ã© chamada de **potÃªncia do sinal**. Para um sinal $x(t)$, definimos sua potÃªncia $P$ por $$P = \lim_{T \rightarrow \infty}\frac{1}{T} \int_{-T/2}^{T/2} |x(t)|^2 dt$$

### Observando a energia e potÃªncia de um sinal que representa um fenÃ´meno fÃ­sico:

Em muitas aplicaÃ§Ãµes, os sinais que consideramos estÃ£o diretamente relacionados a **quantidades fÃ­sicas** e a partir deles pode-se extrair a **potÃªncia ou energia de um sistema fÃ­sico**.  Por exemplo, se $v(t)$ e $i(t)$ sÃ£o, respectivamente, a tensÃ£o e a corrente atravÃ©s de um resistor com resistÃªncia $R$, entÃ£o a potÃªncia instantÃ¢nea Ã© $$p(t) = v(t)i(t) = \frac{1}{R}v^2(t)$$. A energia total dissipada no intervalo de tempo $t_1\le t\le t_2$ Ã© 
$$\int_{t_1}^{t_2}p(t)dt = \int_{t_1}^{t_2}\frac{1}{R}v^2(t)dt $$, e a potÃªncia mÃ©dia durante esse intervalo de tempo Ã© $$\frac{1}{t_2 - t_1}\int_{t_1}^{t_2}p(t)dt = \frac{1}{t_2 - t_1}\int_{t_1}^{t_2}\frac{1}{R}v^2(t)dt$$.

### Sinais DeterminÃ­sticos e AleatÃ³rios

**Sinal DeterminÃ­stico** Ã© aquele cuja descriÃ§Ã£o fÃ­sica Ã© completamente conhecida, seja na forma matemÃ¡tica ou na forma grÃ¡fica.

```python
import numpy as np
import matplotlib.pyplot as plt

fs=100 # freq. amostragem, Hz
fa=95  # freq. sinal analÃ³gico, Hz
T=0.2  # duracao em segundos

t = np.arange(T*fs+1)/fs #tempo discreto
ta = np.arange(T*fs*100)/(fs*100) # tempo

pi = np.pi
sd=np.sin(2*pi*(fs-fa)*t)  # sinal discreto
sa=np.sin(2*pi*fa*ta)      # sinal analÃ³gico

plt.figure(figsize=(12,15))
plt.subplot(311)
plt.plot(ta,sa,color='k'), plt.grid()
plt.title("Sinal analÃ³gico, senoide de %g Hz" % fa)
plt.xlabel("t, s")
plt.subplot(312)
plt.stem(t,sd)
plt.plot(ta,sa,color='k')
plt.plot(t,sd, ':')
plt.grid()
plt.xlabel("t, s")
plt.title("Senoide de %g Hz amostrada a %g Hz" % (fa, fs))
plt.subplot(313)
plt.stem(np.arange(T*fs+1),sd)
plt.title("Sinal discreto")
plt.xlabel("n")
plt.show()
```
