## Sinais PeriÃ³dicos e nÃ£o-periÃ³dicos <a class="anchor" id="section_2_3"></a>

Um sinal $x(t)$ Ã© dito periÃ³dico se para alguma constante positiva $T$ $$x(t) = x(t+T) \;\; \text{para todo t} $$

Em outras palavras, um sinal periÃ³dico tem a propriedade de nÃ£o se modificar pelo deslocamento no tempo de $T$.

* O menor valor de $T$ que satisfaÃ§a a equaÃ§Ã£o Ã© o **perÃ­odo fundamental** do sinal $x(t)$

<div> <img src="figures/periodico.png"/></div> 

```python
def periodic_function(func, period):
    return lambda x: func(((x ) % period )  )

# Taxa de amostragem do sinal que iremos gerar
n = 100
T = 10

# # Amostrando o sinal no tempo
a = np.linspace(0, 30, n, endpoint=False)

sinal = periodic_function(lambda x: x, T)

plt.plot(a,sinal(a))
plt.show()
```

Um sinal que nÃ£o comeÃ§ar antes de $t = 0$ Ã© um **sinal causal**. Em outras palavras, x(t) Ã© um sinal causal se:
$$x(t) = 0 \;\;\; t < 0$$

Sinais Causais:

<div> <img src="figures/causais.png" alt="Drawing" style="width: 600px;"/></div> 

Sinais nÃ£o-causais:

<div> <img src="figures/nao_causais.png" alt="Drawing" style="width: 600px;"/></div> 

```python
# Gera sinal causal com frequÃªncia de F Hz
def sinal_senoidal(periodo, frequencia, tempo):
    sinal = np.sin(periodo*frequencia*tempo)
    return sinal

# Taxa de amostragem do sinal que iremos gerar
n = 1000

# Amostrando o sinal no tempo
x = np.linspace(0, 5, n, endpoint=False)

# Definindo o perÃ­odo e frequencia do sinal
p = 4*np.pi
f = 1

# VizualizaÃ§Ã£o
plt.plot(x, sinal_senoidal(p,f,x))
plt.show()
```

## Energia e potÃªncia de um sinal <a class="anchor" id="section_2_4"></a>

### Tamanho de um Sinal <a class="anchor" id="section_2_4_1"></a>

O tamanho de qualquer entidade Ã© um nÃºmero que indica a largura ou o comprimento da entidade. Genericamente falando, a amplitude do sinal varia com o tempo. **Como um sinal que existe em um certo intervalo de tempo com amplitude variante pode ser medido por um nÃºmero que irÃ¡ indicar o tamanho ou a forÃ§a do sinal?** Tal
medida deve considerar nÃ£o apenas a amplitude do sinal, mas tambÃ©m sua duraÃ§Ã£o. 

Por exemplo, se quisermos utilizar um Ãºnico nÃºmero V como medida do tamanho de um ser humano, devemos considerar nÃ£o somente seu peso, mas tambÃ©m sua altura. Se fizermos uma consideraÃ§Ã£o que a forma da pessoa Ã© um cilindro cuja variÃ¡vel Ã© o raio r (o qual varia com a altura h), entÃ£o uma possÃ­vel medida do tamanho de uma pessoa de altura H Ã© o volume V da pessoa, dado por $$V = \int_o^H r^2(h)dh$$

### Energia de um Sinal <a class="anchor" id="section_2_4_2"></a>

Argumentando desta forma, podemos considerar a Ã¡rea abaixo do sinal $x(t)$ como uma possÃ­vel medida de seu tamanho, pois a Ã¡rea irÃ¡ considerar nÃ£o somente a amplitude, mas tambÃ©m sua duraÃ§Ã£o.$$Tamanho = \int_{-\infty}^{\infty} x(t)dt$$ 


```python
# Taxa de amostragem do sinal que iremos gerar
n = 1000

# Amostrando o sinal no tempo
x = np.linspace(-10, 10, n, endpoint=False)

#FunÃ§Ã£o que calcula o valor da amplitude do sinal para cada instante de tempo amostrado
sinal = np.sin(x)

#vizualizaÃ§Ã£o
plt.rcParams["figure.figsize"] = [7.50, 3.50]

plt.fill_between(x, sinal, step="pre", alpha=0.4)
plt.plot(x, sinal, drawstyle="steps")
plt.xlabel('tempo (t)')
plt.ylabel('Amplitude  x(t)')

plt.show()
```

Entretanto, esta medida ainda Ã© defeituosa, pois mesmo para um sinal grande $x(t)$, suas Ã¡reas positivas e negativas podem se cancelar, indicando um sinal de tamanho pequeno:

Esta dificuldade pode ser corrigida pela definiÃ§Ã£o do tamanho do sinal como a Ã¡rea debaixo de $x^2(t)$ a qual Ã© sempre positiva. Podemos chamar essa medida de **energia do sinal**: $$E = \int_{-\infty}^{\infty} x^2(t)dt\$$

```python

#FunÃ§Ã£o que calcula o valor da amplitude do sinal para cada instante de tempo amostrado
sinal = np.power(np.sin(x),2)

#vizualizaÃ§Ã£o
plt.rcParams["figure.figsize"] = [7.50, 3.50]

plt.fill_between(x, sinal, step="pre", alpha=0.4)
plt.plot(x, sinal, drawstyle="steps")
plt.xlabel('tempo (t)')
plt.ylabel('Amplitude  x(t)')

plt.show()
```

Essa definiÃ§Ã£o pode ser generalizada para um sinal complexo $x(t)$, sendo dada por $$E = \int_{-\infty}^{\infty} |x(t)|^2 dt$$

**A energia do sinal deve ser finita** para que seja uma medida significativa do tamanho do sinal. 
Uma condiÃ§Ã£o necessÃ¡ria para que a energia seja finita Ã© que a amplitude do sinal â 0 quando |t|â â. Caso contrÃ¡rio a integral da equaÃ§Ã£o acima nÃ£o irÃ¡ convergir.

### Potencia de um sinal  <a class="anchor" id="section_2_4_3"></a>

Quando a amplitude do sinal x(t) nÃ£o â 0 quando |t|ââ, a energia do sinal Ã© infinita. 

<div> <img src="figures/s2.png"/></div> 

Uma medida mais significativa do tamanho do sinal neste caso Ã© a energia mÃ©dia, se ela existir. Esta medida Ã© chamada de **potÃªncia do sinal**. Para um sinal $x(t)$, definimos sua potÃªncia $P$ por $$P = \lim_{T \rightarrow \infty} \int_{-T/2}^{T/2} |x(t)|^2 dt$$

### Observando a energia e potÃªncia de um sinal que representa um fenÃ´meno fÃ­sico:

Em muitas aplicaÃ§Ãµes, os sinais que consideramos estÃ£o diretamente relacionados a **quantidades fÃ­sicas** e a partir deles pode-se extrair a **potÃªncia ou energia de um sistema fÃ­sico**.  

Por exemplo, se $v(t)$ e $i(t)$ sÃ£o, respectivamente, a tensÃ£o e a corrente atravÃ©s de um resistor com resistÃªncia $R$, entÃ£o a potÃªncia instantÃ¢nea Ã© 

$$p(t) = v(t)i(t) = \frac{1}{R}v^2(t)$$

A energia total dissipada no intervalo de tempo $t_1\le t\le t_2$ Ã© 
$$\int_{t_1}^{t_2}p(t)dt = \int_{t_1}^{t_2}\frac{1}{R}v^2(t)dt $$

e a potÃªncia mÃ©dia durante esse intervalo de tempo Ã© 

$$\frac{1}{t_2 - t_1}\int_{t_1}^{t_2}p(t)dt = \frac{1}{t_2 - t_1}\int_{t_1}^{t_2}\frac{1}{R}v^2(t)dt$$
